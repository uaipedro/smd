## Modelo de distribuição de espécies (SDM - *Species Distribution Models*)






Um modelo de distribuição de espécies é uma equação que pode ser transformada em um mapa que aponta a **adequação ao habitat** de uma espécie.

**Nota: Não confundir com probabilidade da espécie ser encontrada ou com conceitos mais amplos de nicho.**

Também pode ser chamado de:

*   Modelagem de Envelope Climático (*Climate Envelope Modeling*)
*   Modelagem de Habitat (*Habitat Modeling*)
*   Modelagem de Nicho Ecológico (*Ecological Niche Modeling*)
*   Modelagem de Nicho Ambiental (*Environmental Niche Modeling*)

### Passos para modelar um SDM

1. Coletar as localizações (georeferenciadas) da espécie alvo.
2. Coletar as **variáveis ambientais** dos locais da ocorrência da espécie.
3. Usar as **variáveis ambientais** para treinar um modelo de predição da **adequação ao habitat** em função dessas variáveis.
4. Aplicar a equação a uma área (que faça sentido dentro do escopo) em cada ponto, produzindo o mapa de **adequação ao habitat**.
5. (Opicional) Extrapolar (espacialmente e/ou temporalmente) o modelo para um outro cenário (possível ou previsto).

## Nicho Fundamental
O modelo de distribuição de espécies é baseado no conceito de nicho fundamental:


> O nicho fundamental é o range de ambientes nos quais uma espécie pode manter uma população que se reproduz com frequência suficiente para não encolher. (https://canal.cecierj.edu.br/012016/f319fd50a2fb524c988d45431bea83ab.pdf)

O nicho fundamental não considera interações com outras espécies ou a possível habilidade da espécie se dispersar e reproduzir em outros tipos de ambiente, apenas considera quais conjuntos de condições ambientais uma espécie consegue tolerar.
Não confundir com **nicho realizado**, esse sim representante de onde a espécie realmente está.

## Adequação ao habitat (tradução livre de *Habitat suitability*)

A adequação ao habitat é um indicador do quanto uma espécie viveria bem em um determinado local, caso conseguisse chegar lá. **Não deve ser confundida com a probabilidade de a espécie ser encontrada no local.**


## Variáveis Ambientais / Bioclimáticas


Variáveis Bioclimáticas são medidas indicadoras de condições climáticas / ambientais de um local durante um período, obtidas online através de consórcios como:
* World Clim
* Chelsa
