```python
import pandas as pd
from pygbif import occurrences as occ
import folium
```

```python
def busca_especie_no_gbif(nome_cientifico, size=100):

  infos_interesse = ["taxonKey","genus", "species", "decimalLatitude", "decimalLongitude", "country", "coordinateUncertaintyInMeters", "year", "month", "day"]

  complete = []
  for page in range(int(size/300)+1):

    results = occ.search(scientificName = nome_cientifico, limit=300, offset=300*page, hasCoordinate=True, country="BR").get("results")
    complete = [*complete, *results]

  df = pd.DataFrame(complete)

  return df[infos_interesse]
```

```python
def conta_ocorrencias_especie_gbif(taxon_key):
  return occ.count(taxonKey=taxon_key)
```

```python
conta_ocorrencias_especie_gbif(2436480) # codigo do Leontopithecus chrysopygus
```

```python
busca_especie_no_gbif("Leontopithecus chrysopygus")
```

```python
print(f"Tamanho: {conta_ocorrencias_especie_gbif(2684940)}")
df_araucaria = busca_especie_no_gbif("Araucaria angustifolia*", size=2017)
df_araucaria
```

```python
def cria_mapa(df):
  # Criar um mapa centrado em uma localização inicial
  mapa = folium.Map(location=[df['decimalLatitude'].mean(), df['decimalLongitude'].mean()], zoom_start=5)

  # Adicionar marcadores para cada ponto no DataFrame
  for index, row in df.iterrows():
      folium.Marker(location=[row['decimalLatitude'], row['decimalLongitude']]).add_to(mapa)

  return mapa
```

```python
cria_mapa(df_araucaria)
```
